## Predicting Flower Color  Using KNN

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Dataset
data = {
    'Flower Name': ['Rose', 'Tulip', 'Daisy', 'Rose', 'Lily',
                    'Violet', 'Iris', 'Lavender', 'Orchid', 'Iris',
                    'Daisy', 'Lily', 'Tulip', 'Orchid', 'Violet'],
    'Petal Length': [1.0, 2.0, 2.5, 3.0, 3.2,
                     4.0, 5.0, 5.5, 6.0, 6.3,
                     1.5, 2.8, 3.6, 5.2, 6.7],
    'Petal Width':  [1.0, 1.5, 1.7, 2.0, 2.2,
                     3.0, 3.2, 3.4, 4.0, 4.3,
                     1.2, 1.9, 2.5, 3.3, 4.1],
    'Color': ['Pink', 'Pink', 'Pink', 'Pink', 'White',
              'Violet', 'Violet', 'Violet', 'Violet', 'Violet',
              'White', 'White', 'Pink', 'Violet', 'Violet']
}

df = pd.DataFrame(data)
df.head()
```

```python
# visualisation
# Color map for scatter plot
color_map = {'Pink': 'pink', 'White': 'lightgray', 'Violet': 'purple'}

plt.figure(figsize=(8,6))
for color in df['Color'].unique():
    subset = df[df['Color'] == color]
    plt.scatter(subset['Petal Length'], subset['Petal Width'],
                color=color_map[color], label=color, s=100, edgecolor='black')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Flower Dataset Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prepare features and labels
X = df[['Petal Length', 'Petal Width']]
y = df['Color']
```

```python
# Initialize 
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

```python
# New flower to predict
mystery_flower = [[3.5, 2.1]]
predicted_color = knn.predict(mystery_flower)[0]

print(f"Predicted color for flower with Petal Length = 3.5cm and Width = 2.1cm is: {predicted_color}")

```
